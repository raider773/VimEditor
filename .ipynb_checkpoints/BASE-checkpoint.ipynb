```python
import random

# Definición del problema
colors = ['roja', 'verde', 'blanca', 'amarilla', 'azul']
professions = ['Matemático', 'hacker', 'analista', 'Desarrollador', 'Ingeniero']
languages = ['Python', 'Java', 'JavaScript', 'C#', '']
databases = ['Redis', 'MongoDB', 'HBase', 'Cassandra', 'Neo4J']
editors = ['Vim', 'Brackets', 'Atóm', 'Notepad++', 'Sublime Text']

def generate_individual():
    individual = {
        'color': random.choice(colors),
        'profession': random.choice(professions),
        'language': random.choice(languages),
        'database': random.choice(databases),
        'editor': random.choice(editors)
    }
    return individual

def generate_population(size):
    population = []
    for _ in range(size):
        individual = generate_individual()
        population.append(individual)
    return population

def calculate_fitness(individual):
    fitness = 0
    if individual['profession'] == 'Matemático':
        fitness += 1
    if individual['language'] == 'Python':
        fitness += 1
    if individual['editor'] == 'Vim':
        fitness += 1
    if individual['color'] == 'verde' and individual['editor'] == 'Brackets':
        fitness += 1
    if individual['profession'] == 'analista' and individual['editor'] == 'Atóm':
        fitness += 1
    if individual['color'] == 'blanca' and individual['color'] == 'verde':
        fitness += 1
    if individual['database'] == 'Redis' and individual['language'] == 'Java':
        fitness += 1
    if individual['database'] == 'Cassandra' and individual['language'] == 'C#':
        fitness += 1
    if individual['editor'] == 'Notepad++' and editors.index(individual['editor']) == len(editors) // 2:
        fitness += 1
    if individual['profession'] == 'Desarrollador' and colors.index(individual['color']) == 0:
        fitness += 1
    if individual['database'] == 'HBase':
        index = languages.index(individual['language'])
        if index > 0 and index < len(languages) - 1:
            if individual['language'] == 'JavaScript' or individual['language'] == 'C#':
                fitness += 1
    if individual['database'] == 'Cassandra':
        index = professions.index(individual['profession'])
        if index > 0 and index < len(professions) - 1:
            if individual['profession'] == 'Desarrollador' or individual['profession'] == 'Ingeniero':
                fitness += 1
    if individual['database'] == 'Neo4J' and individual['editor'] == 'Sublime Text':
        fitness += 1
    if individual['profession'] == 'Ingeniero' and individual['database'] == 'MongoDB':
        fitness += 1
    if individual['profession'] == 'Desarrollador' and individual['color'] == 'azul':
        fitness += 1
    return fitness

def evaluate_population(population):
    population_fitness = []
    for individual in population:
        fitness = calculate_fitness(individual)
        population_fitness.append((individual, fitness))
    return population_fitness

def select_parents(population_fitness):
    sorted_population = sorted(population_fitness, key=lambda x: x[1], reverse=True)
    parents = [sorted_population[0][0], sorted_population[1][0]]
    return parents

def crossover(parents):
    crossover_point = random.randint(1, 4)
    child1 = {
        'color': parents[0]['color'],
        'profession': parents[0]['profession'],
        'language': parents[0]['language'],
        'database': parents[0]['database'],
        'editor': parents[1]['editor']
    }
    child2 = {
        'color': parents[1]['color'],
        'profession': parents[1]['profession'],
        'language': parents[1]['language'],
        'database': parents[1]['database'],
        'editor': parents[0]['editor']
    }
    return child1, child2

def mutate(individual):
    attribute = random.choice(['color', 'profession', 'language', 'database', 'editor'])
    if attribute == 'color':
        individual['color'] = random.choice(colors)
    elif attribute == 'profession':
        individual['profession'] = random.choice(professions)
    elif attribute == 'language':
        individual['language'] = random.choice(languages)
    elif attribute == 'database':
        individual['database'] = random.choice(databases)
    elif attribute == 'editor':
        individual['editor'] = random.choice(editors)
    return individual

def genetic_algorithm(population_size, generations):
    population = generate_population(population_size)
    best_solution = None
    best_fitness = 0

    for _ in range(generations):
        population_fitness = evaluate_population(population)
        parents = select_parents(population_fitness)
        child1, child2 = crossover(parents)
        child1 = mutate(child1)
        child2 = mutate(child2)
        new_population = [child1, child2] + [individual for individual, _ in population_fitness[2:]]

        population = new_population     

        best_individual, best_individual_fitness = max(population_fitness, key=lambda x: x[1])
        if best_individual_fitness > best_fitness:
            best_solution = best_individual
            best_fitness = best_individual_fitness

    return best_solution

# Parámetros del algoritmo genético
population_size = 100
generations = 300

# Ejecutar el algoritmo genético
best_solution = genetic_algorithm(population_size, generations)

print("Mejor solución encontrada:")
print(best_solution)
```
